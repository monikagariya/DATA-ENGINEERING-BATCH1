#CSV FILE USING .WRITELINES()

```python
header = ['id','Name','Gender','Age']
data = [[253,'Sonika','Female',18],[254,'Prince','male',30]]
filename ='marksheet.csv'
with open(filename,'w') as file:
    file.write(str(header)+',')
    file.write('n')
    for row in data:
        for x in row:
            file.write(str(x)+', ')
            file.write('n')
               
```

#PYTHON LAMBDA FUNCTION WITH LIST

```python
is_even_list =[lambda arg =x:
               arg *10 
               for x in range(1,5)]

for item in is_even_list:
    print(item())
```

#USING LAMBDA FUNCTION WITH FILTER()

```python
li=[5,7,23,97,54,98,76,73,61]
final_list = list(filter(lambda x: (x%2 !=0), li))
print(final_list)
```

#USING LAMBDA FUNCTION WITH MAP()

```python
li=[5,7,23,97,54,98,76,73,61]
final_list = list(map(lambda x: x*2, li))
print(final_list)
```

#Using LAMBDA WITH REDUCE()

```python
from functools import reduce
li = [5,8,10,20,50,100]
sum = reduce((lambda x, y: x+y),li)
print(sum)
```
